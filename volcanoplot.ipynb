```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('cE14_vs_cGFP.csv')
```

```python
# Set significance threshold (adjust as needed)
alpha = 0.05

# Separate significant and non-significant points
significant = data[data['p_val_adj'] < alpha]
non_significant = data[data['p_val_adj'] >= alpha]

# Create the volcano plot
plt.figure(figsize=(10, 6)
plt.scatter(non_significant['logFC'], -1 * np.log10(non_significant['p_val_adj']), color='gray', alpha=0.5, label='Non-Significant')
plt.scatter(significant['logFC'], -1 * np.log10(significant['p_val_adj']), color='red', label='Significant')

# Label genes with fold change >= 1
for index, row in significant.iterrows():
    plt.text(row['logFC'], -1 * np.log10(row['p_val_adj']), row['gene_names'], fontsize=8, ha='center', va='bottom', color='black')

# Add labels and legend
plt.xlabel('Fold Change')
plt.ylabel('-log10(adj P-value)')
plt.axhline(-1 * np.log10(alpha), color='black', linestyle='--', label=f' = {alpha}')
plt.title('Volcano Plot')
plt.legend()

plt.axvline(1, zorder = 0, c = 'k', lw = 2, ls = '--')
plt.axvline(-1, zorder = 0, c = 'k', lw = 2, ls = '--')
```

```python
#This is the working code

# Set significance threshold (adjust as needed)
alpha = 0.05

# Define your fold change thresholds
fold_change_up = 1
fold_change_down = -1

# Separate Upregulated, No Difference, and Downregulated points
upregulated = data[(data['p_val_adj'] < alpha) & (data['logFC'] >= fold_change_up)]
no_difference = data[data['p_val_adj'] >= alpha]
downregulated = data[(data['p_val_adj'] < alpha) & (data['logFC'] <= fold_change_down)]

# Create the volcano plot
plt.figure(figsize=(10, 6))

# Plot Upregulated, No Difference, and Downregulated points
plt.scatter(upregulated['logFC'], -1 * np.log10(upregulated['p_val_adj']), color='red', label='Upregulated')
plt.scatter(no_difference['logFC'], -1 * np.log10(no_difference['p_val_adj']), color='gray', alpha=0.5, label='No Difference')
plt.scatter(downregulated['logFC'], -1 * np.log10(downregulated['p_val_adj']), color='blue', label='Downregulated')

# Label genes that meet the criteria for Upregulated and Downregulated
for index, row in upregulated.iterrows():
    plt.text(row['logFC'], -1 * np.log10(row['p_val_adj']), row['gene_names'], fontsize=12, ha='center', va='bottom', color='black')

for index, row in downregulated.iterrows():
    plt.text(row['logFC'], -1 * np.log10(row['p_val_adj']), row['gene_names'], fontsize=12, ha='center', va='bottom', color='black')

# Add labels and legend
plt.xlabel('Fold Change')
plt.ylabel('-log10(adj P-value)')
plt.title('Volcano Plot')
plt.legend()

#plt.savefig('volcano_plot.pdf', format='pdf')



# Show the plot
plt.show()
```
